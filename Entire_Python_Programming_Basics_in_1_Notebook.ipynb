## Decorator

Decorator performs additional logic on an existing function and returns a function. it will take a function and return a function.

```python
# use decorator to doubles the result of a function

def doubledecorator(fun):
  def inner_func():
    result = fun()
    return result*2
  return inner_func

def num():
  return 5

resultfun = doubledecorator(num)
resultfun()
```

## using @ for invoking decorator

```python
# use decorator to doubles the result of a function

def doubledecorator(fun):
  def inner_func():
    result = fun()
    return result*2
  return inner_func

@doubledecorator
def num():
  return 5

print(num()) # here you can eliminate the need for assigning the function to a variable and then using the print command on that variable
```

## Decorating to strings

```python
# create a function that takes a name and another function that appends a  string saying how are you + the name
def decorName(fun):
  def inner_fun(name):
    result = fun(name)
    return "How are you doing " + result
  return inner_fun


@decorName
def hello(name):
  return name

print(hello('Aftab'))
```

# Section 15: Inheritence

```python
# understanding inheritence using bmw use case
class BMW:
  def __init__(self, make, model, year):
    self.model = model
    self.make = make
    self.year = year

  def start(self):
    print('Starting a Car')

  def stop(self):
    print('stop the car')


class threeseries(BMW):
  def __init__(self,cruisecontrolenabled,make,model,year): 
    BMW.__init__(self,make,model,year)
    self.cruisecontrolenabled = cruisecontrolenabled

class fiveseries(BMW):
  def __init__(self,parkingassitenabled,make,model,year):  # this class is inheriting 
    BMW.__init__(self,make,model,year)
    self.parkingassitenabled = parkingassitenabled
    
  def display(self):
    print(self.parkingassitenabled)


threeseriescar = threeseries(True,'BMW','328i',2008)
print(threeseriescar.cruisecontrolenabled)
print(threeseriescar.make)
print(threeseriescar.model)
print(threeseriescar.year)
print(threeseriescar.start()) # invoking the method of the parent class because of inheritance
print(threeseriescar.stop())
```

# Section 17 : Abstraction

Abstract method is a method that doesnt have a body. Abstract classes and methods are used mostly in API creation

Abstract class are the classes that has atleast one abstract methods in the class
Abstract methods are those methods which doesnt have body in the method , its just an declaration and we give pass function to pass it simply without performing anything
We need to import module ABC - Abstract Base Classes for using the abstract class method

Abstraction is like "Woh wada jo kattappa ke purkho ne kiya tha ke unki family main har aadmi mahismati singhasan ka ghulam ban ke rahega puri zindagi, So parent classes contracts are to be obliged by child classes"

```python
from abc import ABC,abstractmethod # you need to import this as in python by default you cant define an abstract class and methods

class computer(ABC): # this is an abstract class
  @abstractmethod # this is an abstract method
  def process(self):
    pass

class laptop(computer):
  def process(self):
    print('Its Running')
com1 = laptop()
com1.process()
```

## Excercise for ABC methods and class

```python
from abc import ABC, abstractmethod

class touchscreenlaptop(ABC):

  @abstractmethod
  def scroll(self):
    pass

  @abstractmethod
  def click(self):
    pass

class HP(touchscreenlaptop):
  def scroll(self):
    print('HP Laptop Scrolls')

class Dell(touchscreenlaptop):
  def scroll(self):
    print('Dell laptop scrolls')

class HPnotebook(HP):
  def click(self):
    print("HP click enabled")

class DellNotebook(Dell):
  def click(self):
    print('Dell Click Enabled')


hp = HPnotebook()
de = DellNotebook()
hp.scroll()
hp.click()
de.scroll()
de.click()
```

```python
import threading
# Get the name of the current running thread, which by default has to be main thread
print('Current thread that is running is : ', threading.current_thread().getName())

# to compare current thread to main thread we can do as follows
if threading.current_thread()==threading.main_thread():
  print('Main Thread is running')
else:
  print('Some Other thread is running')
```

## Thread using a function

```python
import threading 
from threading import *
def displaynumber():
  print(current_thread().getName()) # this will give the main thread
  i = 0
  while i<=10:
    print(i)
    i+=10

t = Thread(target=displaynumber) # no need to put () in front of function as only function name is required
t.start() # this will start the thread
print(current_thread().getName()) # this will give the new thread that has spun off
```

## Thread Extending thread class

```python
from threading import Thread

# using super class thread for extending thread class
class mythread(Thread): # Inheriting the thread class
  def run(self): # instead of creating a seperate function create a run function 
    i = 0
    while i<=10:
      print(i)
      i+=1

t = mythread() # assign class to a object
t.start() # instead of invoking the run function start a thread on the same object , else the run function will work on the main thread
```

  # Thread using class

```python
from threading import *

class Mythread:
  def displaynumbers(self): # make sure only to pass self argument and nothing more  
    i = 0
    print(current_thread().getName()) # use print method here to make sure the current thread which is newly created is printed
    while (i<=10): # remember brackets around i<=10 is important else while loop wont print the numbers, though the sequence is executed
      print(i)
      i+=1

obj = Mythread()  # create an object of class my thread
t = Thread(target=obj.displaynumbers) # assign it to thread class using target method
t.start() # start a new thread instead of using main thread
```

## Multi Threading in action

```python
from threading import *
import time


def calc_cube(numbers):
  for i in numbers:
    time.sleep(0.2)
    print('cube : ', i**3)

def calc_square(numbers):
  for i in numbers:
    time.sleep(0.2)
    print('square : ', i**2)

arr = [2,3,4,5]

t = time.time() # start monitoring time

t1= Thread(target=calc_cube, args=(arr,)) # thread using a funciton
t2= Thread(target=calc_square, args=(arr,)) # Thread using a function

t1.start() # start the thread 1
t2.start() # start the thread 2

t1.join() # wait for thread 1 to complete
t2.join() # wait for thread 2 to complete


print('Done in : ', time.time()-t) # print total time taken
```
