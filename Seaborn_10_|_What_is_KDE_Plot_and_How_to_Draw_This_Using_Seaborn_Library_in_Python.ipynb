1. Create basic KDE plot for Total column

```python
sns.kdeplot(data=mart, x='Total')
```

2. Create KDE plot for all numeric variables in the dataframe

```python
sns.kdeplot(data=mart)
```

3. Adjust the smoothing using bw_adjust

```python
sns.kdeplot(data=mart, x= 'Total', bw_adjust=0.2)
```

4. Group the KDE on a category variable

```python
sns.kdeplot(data=mart, x= 'Total', bw_adjust=1, hue='Payment')
```

5. Stack KDE on a category using Multiple arguement

```python
sns.kdeplot(data=mart, x= 'Total', bw_adjust=1, hue='Payment', multiple='stack')
```

6. Use log scaling to map variable on the KDE plot


```python
sns.kdeplot(data=mart, x= 'Total', bw_adjust=1, hue='Payment', multiple='stack', log_scale=True)
```

7. Change styling of the hued KDE, using line widht, palette , alpha etc

```python
sns.kdeplot(data=mart, x= 'Total', bw_adjust=1, hue='Payment', multiple='stack', log_scale=True, palette='husl',linewidth=2, alpha=0.8)
```

8.  Create  a bivariate KDE plot

```python
sns.kdeplot(data=mart, x= 'Unit price', y='gross income', bw_adjust=1, multiple='stack', log_scale=False, palette='husl',linewidth=2, alpha=0.8)
```

9. Group the bivariate KDE on category variable

```python
sns.kdeplot(data=mart, x= 'Unit price', y='gross income', bw_adjust=1,hue='Gender', multiple='stack', log_scale=False, fill=True, palette='husl',linewidth=2, alpha=0.8, levels=5,thresh=.2)
```
