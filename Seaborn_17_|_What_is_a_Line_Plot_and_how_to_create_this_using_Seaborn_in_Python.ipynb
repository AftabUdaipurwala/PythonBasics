# ***Libraries and Dataset loading***

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
from scipy import stats
import warnings
warnings.filterwarnings("ignore")
from google.colab.patches import cv2_imshow
import os

drive.mount('/content/drive/')
!ls '/content/drive/MyDrive/Data Science Learning Path/Python Seaborn Library/'
mart = pd.read_csv('/content/drive/MyDrive/Data Science Learning Path/Python Seaborn Library/train.csv')
mart['Tier'] = mart['Outlet_Location_Type'].str.strip().str[-1]
mart.head(5)
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales')
```

2. Test with different confidence interval

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator='mean')
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', ci=None, estimator=None)
```

3. Group using Hue and Different Palette

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', palette='husl', ci='sd', n_boot=1,hue='Outlet_Location_Type')
```

4. Group using style and different marker and dashes

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', palette='husl', ci=None, style='Outlet_Size',markers=True,dashes=False)
```

5. Group using sizes

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', palette='husl', ci=None, style='Outlet_Size',markers=True,dashes=False, size='Outlet_Size', sizes=(0.5,5), color='red')
```

6. Use different semantic styling parameters

```python
plt.figure(figsize=(15,7))
sns.lineplot(data=mart, x='Outlet_Year', y='Sales', palette='husl', ci=None, hue='Outlet_Size', style='Outlet_Size',  markers=True,dashes=False, size='Outlet_Size', sizes=(0.5,5))
```
