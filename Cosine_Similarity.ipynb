```python
import pandas as pd
import numpy as np
from sklearn  import metrics
from scipy import spatial
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# ***Cosine Similarity***

```python
List1 = [4, 47, 8, 3]
List2 = [3, 52, 12, 16]
result = 1 - spatial.distance.cosine(List1, List2)
print(result)
```

# ***Cosine Distance***

```python
List1 = [4, 47, 8, 3]
List2 = [3, 52, 12, 16]
result = spatial.distance.cosine(List1, List2)
print(result)
```

```python
# Creating a dataframe and doing the cosine similarity
df = pd.DataFrame(
    [{'iphone':3, 'galaxy':1},
    {'iphone':2, 'galaxy':0},
    {'iphone':1, 'galaxy':3},
    {'iphone':1, 'galaxy':2}],
    index=[
           'doc1',
           'doc2',
           'doc3',
           'doc4'
    ]
)
```
