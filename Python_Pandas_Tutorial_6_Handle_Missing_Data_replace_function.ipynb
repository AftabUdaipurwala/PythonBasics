```python
# Read data from a Csv file format
df = pd.read_csv('/content/drive/MyDrive/py-master/pandas/6_handling_missing_data_replace/weather_data.csv', parse_dates=['day']) # parse dates as date while reading the file
df
```

```python
# replace function to replace certain values
new_df = df.replace([-88888,-99999],np.NaN) # we can supply list of values
new_df
```

```python
# replace based on specific column
new_df = df.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.NaN)
new_df
```

```python
# mapping based replacement by creating dictionary
# replace based on specific column and specific values
new_df = df.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},{'temperature':df.temperature.median(),'windspeed':df.windspeed.median(),'event':np.NaN})
new_df
```

```python
# replacement using regex

new_df = df.replace({'temperature':['A-Za-z'],'widnspeed':['A-Za-z']},regex=True)
new_df
```

```python
# replace 1 list by another list with dictionary mapping
new_df = df.replace({'event':['Rain','Snow','Sunny',0]},{'event':[1,2,3,0]})
new_df
```
