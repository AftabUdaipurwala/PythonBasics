# Data Section

```python
temperature_df= pd.DataFrame({
  'city':['Mumbai','Delhi','Chennai','Kolkata','Jaipur'],
    'temperature':[32,45,43,33,41]  
} )
temperature_df
```

```python
wind_speed_df = pd.DataFrame({
    'city':['Mumbai','Delhi','Chennai','Kolkata','Banglore'],
    'windspeed':[2,5,3,8,0]
}) 
wind_speed_df
```

# Inner Join

image.png

```python
# Innerjoin
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city')
new_df
```

# Outer Join

image.png

```python
#outer join
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city', how='outer')
new_df
```

#Left join

image.png

```python
#Left join
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city', how='left')
new_df
```

# Right join

image.png

```python
#right join
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city', how='right')
new_df
```

# Indicator 

```python
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city', how='outer',indicator=True)
new_df
```

# Suffixes


```python
#outer join
new_df = pd.merge(temperature_df,wind_speed_df, on = 'city', how='outer', indicator=True,suffixes=['_temp','_windspeed'])
new_df
```
