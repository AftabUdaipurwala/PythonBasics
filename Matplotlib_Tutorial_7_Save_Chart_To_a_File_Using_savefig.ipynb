```python
val=[100,145,321,200]
exp=['rent','lght','grocer','misc']
```

```python
plt.pie(val,labels=exp, autopct='%.1f%%', shadow=True, explode=[.2,.2,0,0])
plt.legend()
plt.show()
```

```python
plt.pie(val,labels=exp, autopct='%.1f%%', shadow=True, explode=[.2,.2,0,0])
plt.legend()
plt.savefig('Pie Chart', bbox='tight', pad_inches=2, transparent=True)
```
