```python
# Converting multiple datetime formats in a single fomrat we can use the to_datetime fuction in python
dt= ['2017-01-05','Jan 5, 2017', '01/05/2017','2017.01.05','2017/01/05']
pd.to_datetime(dt)
```

```python
# now lets consider there are some garbage values in the list, so by default there will be an exception raised, to take care of it we can do
dt= ['2017-01-05','Jan 5, 2017', '01/05/2017','2017.01.05','2017/01/05','abc']
pd.to_datetime(dt, errors ='ignore') # ignore will not convert the entire list
```

```python
# now lets consider there are some garbage values in the list, so by default there will be an exception raised, to take care of it we can do
dt= ['2017-01-05','Jan 5, 2017', '01/05/2017','2017.01.05','2017/01/05','abc']
pd.to_datetime(dt, errors ='coerce') # coerce will convert garbage to NAT which means not a timeformat
```

```python
# by default the timestamp style in the python is in US format, lets say u want to convert it in the Europe Format
dt= ['2017-01-05','Jan 5, 2017', '01/05/2017','2017.01.05','2017/05/01','abc']
pd.to_datetime(dt, errors ='coerce', dayfirst=True) # converts US format to European format
```
