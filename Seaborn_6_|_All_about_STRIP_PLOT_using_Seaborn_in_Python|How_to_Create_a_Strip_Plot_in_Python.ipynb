Create basic Strip plot

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart, palette='husl',y='Total', )
```

Expand markers in strip plot using jitters

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart, palette='husl',y='Total', jitter=1.1)
```

Draw line around the points using line width

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart,x='Product line', palette='husl',y='Total',linewidth=1)
```

include third categorical variable using hue

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart,x='Product line', palette='husl',y='Total',linewidth=0, hue='Gender')
```

seperate each level of hue using dodge



```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart,x='Product line', palette='husl',y='Total',linewidth=0, hue='Gender', dodge=True)
```

draw strips on top of a voilin plot

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart,x='Product line', palette='husl',y='Total', jitter=0.1)
sns.violinplot(data=mart,x='Product line',y='Total', color='gray')
```

Draw strips on top of a box plot

```python
plt.figure(figsize=(15,5))
sns.stripplot(data=mart,x='Product line', palette='husl',y='Total', jitter=0.1)
sns.boxplot(data=mart,x='Product line',y='Total', color='gray')
```
