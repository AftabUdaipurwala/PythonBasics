# ***2. Advantage its fast and convienent***

```python
SIZE=1000000
l1=range(SIZE)
l2=range(SIZE)

a1=np.arange(SIZE)
a2=np.arange(SIZE)

start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
end= time.time()
print("python list took :", (end-start)*1000)


```

```python
start = time.time()
result = a1+a2
end= time.time()
print("python list took :", (end-start)*1000)
```
