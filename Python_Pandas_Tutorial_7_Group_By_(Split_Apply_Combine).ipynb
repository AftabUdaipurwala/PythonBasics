```python
# for first question using group by function
g = df.groupby(by='city')
for city,city_df in g:
  print(city)
  print(city_df)
  
```

```python
 #to get average windspeed by city
df1 = df[['windspeed','city']]
df1.groupby(by='city').mean()
```

```python
# to plot graphs of each group
import matplotlib.pyplot as plt
%matplotlib inline
df1.groupby('city').plot()
```
