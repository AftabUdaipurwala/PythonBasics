# Data Section

```python
india_weather = pd.DataFrame({
    'city':['Mumbai','Delhi','Chennai','Kolkata'],
    'temperature':[32,45,43,33],
    'humidity':[100,20,40,50]
})
india_weather
```

```python
US_weather = pd.DataFrame({
    'city':['Chicago','orlando','newyork','washington'],
    'temperature':[30,41,41,31],
    'humidity':[10,30,45,54]
})
US_weather
```

# Concat Function

```python
new_df = pd.concat([india_weather,US_weather], ignore_index=True) # ignore index will reset the index in the new table esle the orginal indexes will repeated
new_df
```

```python
new_df = pd.concat([india_weather,US_weather], keys=['India','US']) # to get the cities in the india and US, but make sure ignore_index is not marked false to use this function
new_df
```

```python
new_df.loc['US'] # to get the subset of the dataframe based on the category of the original dataframe
```

```python
temperature_df= pd.DataFrame({
  'city':['Mumbai','Delhi','Chennai','Kolkata'],
    'temperature':[32,45,43,33]  
}, index=[0,1,2,3])# providing index is necessary to make sure that the dataframes are merged properly
temperature_df
```

```python
wind_speed_df = pd.DataFrame({
    'city':['Mumbai','Delhi','Chennai','Kolkata'],
    'windspeed':[2,5,3,8]
}, index=[0,1,2,3]) # providing index is necessary to make sure that the dataframes are merged properly
wind_speed_df
```

```python
#concat 2 dataframes based on the index numbers
pd.concat([wind_speed_df,temperature_df],axis=1)
```

```python
# append series in the dataframe
event = pd.Series(['Humid','Dry','Rain'], name='events')
new_df= pd.concat([india_weather,event],axis=1)
new_df
```
