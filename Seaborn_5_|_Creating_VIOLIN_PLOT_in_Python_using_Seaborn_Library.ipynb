# Create basic violin plot

```python
plt.figure(figsize=(5,5))
sns.violinplot(y='Total',data=mart)
```

# Create Violin plot on 2 categorical one numeric variable and split it using the gender variable

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Product line',hue='Gender',y='Total',data=mart, palette='husl', split=True)
```

# change the box in violin plot to horizontal lines


```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Product line',hue='Gender',y='Total',data=mart, palette='husl', split=True, inner='quartile')
```

# Draw line for each observations in the violin plot


```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Product line',hue='Gender',y='Total',data=mart, palette='husl', split=True, inner='stick')
```

# change amount of smoothing using bw attribute, bw means bandwidht

```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Product line',hue='Gender',y='Total',data=mart, palette='husl', split=True, inner='quartile', bw=2)
```

# cut out the extreme values


```python
plt.figure(figsize=(15,5))
sns.violinplot(x='Product line',hue='Gender',y='Total',data=mart, palette='husl', split=True, inner='quartile', cut=0)
```
