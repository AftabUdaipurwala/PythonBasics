```python
# to create cross tab
pd.crosstab(df.Nationality, df.Handedness)
```

```python
# alongwith totals
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

```python
# multiple variables
pd.crosstab(df.Sex, [df.Nationality,df.Handedness]) # simply pass an array
```

```python
# get percentages in cross tab
pd.crosstab(df.Sex, df.Handedness, normalize='index')
```

```python
# get average age in cross tab
pd.crosstab(df.Sex, df.Handedness, values=df.Age, aggfunc=np.average)
```
