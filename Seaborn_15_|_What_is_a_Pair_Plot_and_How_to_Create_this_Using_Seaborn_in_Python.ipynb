```python
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.columns
```

1. Create basic pair plot

```python
sns.pairplot(data=iris)
```

2. Changing diagonal plot kind to KDE hist or none

```python
sns.pairplot(data=iris, diag_kind='kde')
```

3. Changing the non diagonal plot kind to scatter, kde, hist or reg

```python
sns.pairplot(data=iris, diag_kind='kde', kind='reg')
```

4. Adding hue to pair plot

```python
sns.pairplot(data=iris, diag_kind='kde', kind='reg', hue='species', palette='husl')
```

5. Creating pair plot to specific list of variables

```python
sns.pairplot(data=iris, hue='species', palette='husl',x_vars=['sepal_length', 'sepal_width', 'petal_length'], y_vars=['sepal_width','petal_length'], diag_kind=None)
```

6. Showing lower triangle using corner arguement

```python
sns.pairplot(data=iris, diag_kind='kde', kind='reg', hue='species', corner=True)
```

7. Making changes to specific elements of diagonal and non diagonal elements

```python
sns.pairplot(data=iris, diag_kind='kde', kind='reg', corner=True, plot_kws={'marker':'+'}, diag_kws={'color':'red'})
```

8. Creating KDE plot on top of pair plot 

```python
l=sns.pairplot(data=iris, hue='species', palette='husl')
l.map_upper(sns.kdeplot)
```
