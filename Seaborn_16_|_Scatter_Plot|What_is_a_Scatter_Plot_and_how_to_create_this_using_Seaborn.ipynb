1. Create a basic scatter plot

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales')
```

2. Group by Outlet size

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales', hue='Outlet_Size')
```

3. Group by categorical variable using stlying and markers customization

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales', hue='Outlet_Size', markers= True, style='Outlet_Location_Type' )
```

4. Grouping on basis of hue and use of palette

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales', hue='Outlet_Size', markers= True, style='Outlet_Location_Type', palette='husl' )
```

5. Grouping numeric variable using hue and size

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales', hue='Outlet_Size', size='Tier', sizes=(2,200), legend='full', palette='husl' )
```

6. Changing marker size using s

```python
sns.scatterplot(data=mart, x='Item_MRP',y='Sales',s=500, color='red', edgecolor='blue')
```
