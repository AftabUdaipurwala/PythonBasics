1. Create a basic hist plot on x and y axis

```python
sns.histplot(x='Total', data=mart, palette='Set2')
```

```python
sns.histplot(y='Total', data=mart, palette='Set2')
```

2. Change bin widht using binwidht argument

```python
sns.histplot(x='Total', data=mart, palette='hus;',binwidth=150)
```

3. Change Number of Bin and interval

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, palette='hus;',bins=np.arange(0,1100,50)) # each hist bar shows 50 ka step
plt.xticks(np.arange(0,1100,50))
```

4. Combine with KDE plot

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, palette='husl', kde=True) # each hist bar shows 50 ka step
```

5. Use categorical variable Hue and Stack in multiple arguments

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, palette='husl', hue='Gender',multiple='stack', kde=True) # each hist bar shows 50 ka step
```

6. Make it a step ploy plot using element argument and change the fill

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, palette='husl', hue='Gender',multiple='stack',element='step',fill=False, kde=True) # each hist bar shows 50 ka step
```

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, palette='husl', hue='Gender',multiple='stack',element='poly',fill=True, kde=True) # each hist bar shows 50 ka step
```

7. Use categorical variable and shrink it

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Payment', data=mart, shrink=0.7,  ) # each hist bar shows 50 ka step
```

8. Changing stat from count to probability


```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total', data=mart, stat='probability'  ) # each hist bar shows 50 ka step
```

9. Bivariate Histogram

```python
plt.figure(figsize=(15,5))
sns.histplot(x='Total',y='gross income', data=mart ) # each hist bar shows 50 ka step
```
