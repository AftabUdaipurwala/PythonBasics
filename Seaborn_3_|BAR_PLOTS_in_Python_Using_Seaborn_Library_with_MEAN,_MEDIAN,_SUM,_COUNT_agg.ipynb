**Create basic barplot**

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', data=mart, errwidth=5, errcolor='.7', ci=90,)
```

**Adding Hue to bargraph**

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', data=mart, errwidth=5, errcolor='.7', ci=90, hue='Gender') # bar graph by default shows average values
```

# **Horizontal bar graph**

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci=90, hue='Gender') # bar graph by default shows average values
```

# **Plot the bars in a given order**

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci=90, order = mart['Product line'].sort_values().unique(), hue='Gender',hue_order=['Male','Female']) # Sorted by alphabetical order
```

# **Add Cap and Error Bar**

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci=90, order = mart['Product line'].sort_values().unique(),
            capsize=0.2) # Sorted by alphabetical order
```

# **Remove Error bar**

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            ) 
```

# ***change Bar colour using colour attribute***

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            color='Aquamarine') 
```

# ***Change Colour palette***

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            palette='husl') 
```

# **Using Saturation parameter**

```python
plt.figure(figsize=(5,5))
sns.barplot(y='Product line',x='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            palette='husl', saturation =.1) 
```

# ***Change default aggregation parameter***

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            palette='husl',estimator=sum) 
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Product line',y='Total', data=mart, errwidth=5, errcolor='.7', ci= None, order = mart['Product line'].sort_values().unique(),
            palette='husl',estimator=np.median)  # using median
```
