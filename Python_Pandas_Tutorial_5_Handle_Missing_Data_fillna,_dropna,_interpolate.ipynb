```python
# Read data from a Csv file format
df = pd.read_csv('/content/drive/MyDrive/py-master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv', parse_dates=['day']) # parse dates as date while reading the file
df
```

```python
# set the day as index
df.set_index('day', inplace=True)
```

```python
df
```

# Fillna method

```python
fillby0 = df.fillna(0)
fillby0
```

```python
fillcustom = df.fillna({
    'temperature':df.temperature.mean(),
    'windspeed': df.windspeed.median(),
    'event': 'No event'
})
fillcustom
```

```python
# forward fill and backward fills
df.fillna(method='ffill', axis = 'rows' , limit =2)
```

# Interpolate method

```python
# interpolating the values
df.interpolate()
```

```python
# interpolating with importance given to time 
df.interpolate(method='time') # check the value of temperature in second row with respect to table above its changed from 30 to 29
```

# Drop NA 

```python
# to remove only those rows which are fully empty
df.dropna(how='all')
```

```python
df.dropna(thresh=1)
```

```python
# inserting dates in the index value
dt = pd.date_range('01-01-2017','01-11-2017')
idx= pd.DatetimeIndex(dt)
df = df.reindex(idx)
df
```
