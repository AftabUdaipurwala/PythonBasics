1. Create a basic rug plot

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price')
```

2. Create a rug plot for 2 variables

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income')
```

3. Group it by categorical variable hue

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Payment')
```

4. Change the hieght of the rugs in the plot

```python
plt.figure(figsize=(15,5))
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Payment', height=0.1)
```

5. Combine with KDE plot

```python
plt.figure(figsize=(15,5))
sns.kdeplot(data=mart, x='Unit price',y='gross income', fill=True, hue='Payment')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Payment', height=0.1)
```

6. Combine with Scatter plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=mart, x='Unit price',y='gross income', hue='Payment')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Payment', height=0.1)
```

7. Show Rugs outside of the plot

```python
plt.figure(figsize=(15,5))
sns.scatterplot(data=mart, x='Unit price',y='gross income', hue='Payment')
sns.rugplot(data=mart, x='Unit price', y='gross income', hue='Payment', height=-0.01, clip_on=False)
```
