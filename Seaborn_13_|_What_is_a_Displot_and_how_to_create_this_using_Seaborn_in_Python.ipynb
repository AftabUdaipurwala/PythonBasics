# ***Libraries and Dataset loading***

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
from scipy import stats
import warnings
warnings.filterwarnings("ignore")
from google.colab.patches import cv2_imshow
import os

drive.mount('/content/drive/')
!ls '/content/drive/MyDrive/Data Science Learning Path/Python Seaborn Library/'
mart = pd.read_excel('/content/drive/MyDrive/Data Science Learning Path/Python Seaborn Library/supermarket_sales.xlsx')
mart.head(5)
```

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':2})
```

2. Add Categorical variable

```python
sns.displot(data=mart, x='Total',hue='Payment',multiple='stack', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':2})
```

3. For more than 1 categorical variable

```python
sns.displot(data=mart, x='Total',hue='Payment',col='Gender',multiple='stack', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':2})
```

4. Plot row wise 3rd categorical variable

```python
sns.displot(data=mart, x='Total',hue='Payment',col='Branch',row='Gender',multiple='stack', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':2})
```

5. Make it a seperate bar plot 

```python
sns.displot(data=mart, y='Total',hue='Payment',col='Branch',row='Gender',multiple='stack', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':2})
```

6.  Change type of the plot

```python
sns.displot(data=mart, x='Total',hue='Payment',col='Branch',row='Gender', kind='ecdf')
```
