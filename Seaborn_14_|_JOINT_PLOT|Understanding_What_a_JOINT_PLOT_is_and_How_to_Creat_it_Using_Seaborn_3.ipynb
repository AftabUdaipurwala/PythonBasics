1. Draw a basic Joint plot

```python
sns.jointplot(data=mart,x='Total', y='gross income')
```

2. Change its kind to 'scatter', 'kde', "hist", 'hex','resid','reg'

```python
sns.jointplot(data=mart,x='Total', y='gross income', kind='hist')
```

```python
sns.jointplot(data=mart,x='Total', y='gross income', kind='reg')
```

```python
sns.jointplot(data=mart,x='Total', y='gross income', kind='resid')
```

```python
sns.jointplot(data=mart,x='Total', y='gross income', kind='hex')
```

3. Grouping based on categorical variable

```python
sns.jointplot(data=mart,x='Total', y='gross income', hue='Payment')
```

4. Formatting all individual plot

```python
sns.jointplot(data=mart,x='Total', y='gross income', hue='Payment', palette='husl')
```

```python
sns.jointplot(data=mart,x='Total', y='gross income', joint_kws={'marker':'*', 'color':'red'}, marginal_kws={'color':'green','kde':'True'})
```

5. Adjust hieght, ratio, space and show hide marginal ticks

```python
sns.jointplot(data=mart,x='Total', y='gross income', height = 10, ratio=2,space=2, marginal_ticks=True )
```

6. plot KDE and Rug  plot on top of joint plot


```python
l= sns.jointplot(data=mart,x='Total', y='gross income', color='red')
l.plot_joint(sns.rugplot)
```
