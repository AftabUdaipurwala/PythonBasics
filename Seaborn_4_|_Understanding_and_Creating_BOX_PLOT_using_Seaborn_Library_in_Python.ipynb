# ***Test example***

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line',y='Quantity',data=mart,hue='Gender', palette='husl')
```

Create box plot with one numeric variable

```python
plt.figure(figsize=(5,2))
sns.boxplot(x='Quantity',data=mart, palette='husl', width=0.5)
```

Create box plot with one numeric variable and one categorical variable

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line',y='Quantity',data=mart, palette='husl')
```

Create box plot with one numeric variable and two categorical variable 2nd one as hue

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line',y='Quantity',data=mart,hue='Gender', palette='husl')
```

Adding mean marker using Showmeans attribute and change its style using meanprops

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line',y='Quantity',data=mart,hue='Gender', palette='husl', showmeans=True, meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"yellow","markersize":"15"})
```

change palette and line widht etc

```python
plt.figure(figsize=(15,5))
sns.boxplot(x='Product line',y='Quantity',data=mart,hue='Gender', palette='Set2', linewidth=1)
```

Create box plot for each of the numeric variable in box plot

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=mart, palette='husl')
```
