```python
# Reading from an dictionary
weather_data = {
    'day':['2017-01-01','2017-01-02','2017-01-03'],
    'temperature':[32,35,26],
    'windspeed':[6,7,6],
    'event':['Rain','Sunny','Snow']
}
df = pd.DataFrame(weather_data)
df
```

```python
# Reading from a tuple
weather_data =[
               ('2017-01-01',32,6,'Rain'),
               ('2017-01-02',35,7,'Sunny'),
               ('2017-01-03',26,6,'Snow')
]
df = pd.DataFrame(weather_data, columns=['day','temperature','windspeed','event'])
df
```

```python
# Reading from a list of dictionary 
weather_data =[
               {'day':'2017-01-01','temperature':32,'windspeed':6,'event':'Rain'},
               {'day':'2017-01-02','temperature':35,'windspeed':7,'event':'Sunny'},
               {'day':'2017-01-03','temperature':26,'windspeed':6,'event':'Snow'}
]
df = pd.DataFrame(weather_data)
df
```
