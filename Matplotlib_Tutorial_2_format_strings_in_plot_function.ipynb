```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[51,54,67,34,56,50,48]
```

```python
plt.plot(x,y,'g--.')
```

```python
plt.plot(x,y,'rD')
```

```python
plt.plot(x,y, color='red', marker='D', linestyle='dotted', markersize=20)
```

```python
plt.plot(x,y, color='#C1D7D9', marker='D', linestyle='dotted')
```

```python
plt.plot(x,y, color='maroon', marker='D', linestyle='dotted', alpha=0.5)
```
