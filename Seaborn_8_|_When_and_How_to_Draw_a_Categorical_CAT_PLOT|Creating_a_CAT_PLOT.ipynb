Plot basic cat plot

```python
sns.catplot(x='Payment', y='Total',data=mart, kind='box',hue='Gender', palette='husl')
```

```python
sns.catplot(x='Payment', y='Total',data=mart, kind='box',hue='Gender', col='Product line',row='Customer type', palette='husl')
```

```python
sns.catplot(x='Payment', y='Total',data=mart, kind='violin',hue='Gender', col='Product line',row='Customer type', palette='husl')
```

```python
sns.catplot(x='Payment', y='Total',data=mart, kind='swarm',dodge=True,hue='Gender', col='Product line',row='Customer type', palette='husl')
```
