```python
martpiv = mart.pivot_table(index='Outlet_Year', columns='Outlet_Size',values='Sales')
martpiv
```

```python
sns.set_style('white')
plt.figure(figsize=(15,10))
sns.heatmap(data=martpiv, annot=True, fmt='.0f', annot_kws=dict(size=16,weight='bold'), linewidths=6, linecolor='black', cmap='gnuplot2')
```

```python
martcor= mart.corr()
martcor
```

```python
sns.heatmap(data=martcor, vmax=1,vmin=-1,center=0, cmap='cool_r', annot=True, fmt='.1f', annot_kws=dict(size=16, weight='bold'))
```
