```python
# stacking 2 arrays togather
b=np.arange(6).reshape(3,2)
c=np.arange(6,12,1).reshape(3,2)
```

```python
np.vstack((b,c))
```

```python
np.hstack((b,c))
```

```python
d=np.arange(30).reshape(2,15)
```

```python
np.hsplit(d,3) # vertically slicing in 3 different arrays
```
