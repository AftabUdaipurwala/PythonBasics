```python
import numpy as np
```

```python
a= np.arange(12).reshape(3,4)
a
```

```python
for i in a:
  for cell in i:
      print(cell)
```

```python
for i in a.flatten():
      print(i)
```

```python
# using the same thing using nditer funciton
for x in np.nditer(a,order='C'):
  print(x)
```

image.png

```python
# using the same thing using nditer funciton
for x in np.nditer(a,order='F'):
  print(x)
```

```python
# using the same thing using nditer funciton, using flags function
for x in np.nditer(a,order='F', flags=['external_loop']):
  print(x)
```

```python
# using the same thing using nditer funciton, using flags function
for x in np.nditer(a,order='F', op_flags=['readwrite']):
  x[...]=x*x
```
